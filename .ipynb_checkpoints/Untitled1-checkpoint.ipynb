```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing the data

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head() #viewing the head of the data
```

### Visualizing the data

```python
sns.pairplot(data=df, diag_kind="kde")  #visualizing the data to see any relevant relationships
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();
```

```python
X = df.drop("sales", axis=1)  # spliting the data into the required labels
y=df['sales']
```

```python
X   #viewing the selected labels
```

```python
y
```

## Train Test split under sklearn

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=101)  #splitting the data into  train and test set
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

#### Fitting a linear regression model to the data

```python
from sklearn.linear_model import LinearRegression  #importing the linear regression class
```

```python
model = LinearRegression()  # creating an instance of the linear regression model
```

```python
model.fit(X_train, y_train)  # actual fitting of the model using the training dataset
```

```python
predicted_sales = model.predict(X_test)
predicted_sales
```

```python
import pickle

# Save the trained LSTM model to a pickle file
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
```

### Model Evaluation

```python
# importing the necessary performance metrics to use in evaluation
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
MSE = mean_squared_error(y_test, predicted_sales)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
MAE = mean_absolute_error(y_test, predicted_sales)
MAE
```

```python
mean_test_sales = np.mean(y_test)  #computing the mean of the test sales values
mean_test_sales
```

```python
#expressing the model performance as a percentage
model_performance = (1 - RMSE/mean_test_sales)*100
model_performance
```

#### Discussion
The value obtained above shows that the performance of the model is at 89.19% which is actually a fairly good value.

```python
model_residuals = y_test - predicted_sales
model_residuals
```
